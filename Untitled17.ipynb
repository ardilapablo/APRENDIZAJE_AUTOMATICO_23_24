En este paso del proceso, se busca abordar la presencia de "missing values" en la columna "people vaccinated" a través de un método específico. Se comienza por convertir la columna "date" al tipo datetime, lo que es fundamental para futuros cálculos y comparaciones temporales. Posteriormente, se crea una nueva columna denominada "corrected_people_vaccinated" para guardar los valores corregidos de "people vaccinated" que serán calculados.

Para cada fila con un valor faltante en "people vaccinated", se procede a filtrar el DataFrame para obtener los datos del mismo país y fechas anteriores. Se asegura la consideración de los últimos 3 días (o menos si no están disponibles) utilizando un método de selección y ordenamiento. A continuación, se encuentran y promedian los valores de "people vaccinated" de los días anteriores del mismo país.

Si se encuentran valores anteriores del mismo país, se calcula la media de dichos valores y se actualiza el "people vaccinated" corregido en la fila correspondiente. Es importante destacar que se emplea el redondeo para garantizar la integridad de los datos y el valor corregido. Si no se encuentran suficientes datos anteriores del mismo país para realizar la corrección, se registra esta información a través de mensajes impresos.

```python
# Convertir la columna "date" a tipo datetime
df['date'] = pd.to_datetime(df['date'])

# Crear una nueva columna para almacenar los valores corregidos
df['corrected_people_vaccinated'] = df['people_vaccinated']

# Iterar a través de las filas con NA y realizar la corrección
for index, row in df.iterrows():
    if pd.isnull(row['people_vaccinated']):  # Verificar si el valor es NA
        # Filtrar el DataFrame para obtener solo los datos del mismo país y fechas anteriores
        previous_dates = df[(df['country'] == row['country']) & (df['date'] < row['date'])]['date']

        if not previous_dates.empty:
            # Asegurarse de que consideramos solo los últimos 3 días o menos si no están disponibles
            previous_dates = previous_dates.sort_values(ascending=False).head(3)
            # Encontrar los valores de "people_vaccinated" de los días anteriores del mismo país
            previous_vaccinated = df[(df['country'] == row['country']) & (df['date'].isin(previous_dates))]['people_vaccinated']

            # Calcular la media de los valores existentes y actualizar el valor corregido
            if not previous_vaccinated.empty:  # Verificar si hay valores anteriores del mismo país
                mean_previous_vaccinated = previous_vaccinated.mean()
                rounded_mean = np.round(mean_previous_vaccinated)
                df.at[index, 'people_vaccinated'] = rounded_mean  # Actualizar el valor corregido
                print(f"Fecha: {row['date']} - Valor corregido: {rounded_mean}")
            else:
                print(f"No hay suficientes datos anteriores del mismo país para corregir la fecha: {row['date']}, país: {row['country']}")
        else:
            print(f"No hay suficientes datos anteriores del mismo país para corregir la fecha: {row['date']}, país: {row['country']}")




# Utiliza el método isna() para identificar los valores faltantes y luego usa el método sum() para contarlos
missing_values = df.isna().sum()

# Muestra los valores faltantes en cada columna
print(missing_values)
```

En aras de abordar los desafíos relacionados con los nombres de países que comienzan con "NA" y que no se corrigen con el código inicialmente implementado, se ha optado por realizar un análisis de la matriz de correlación entre variables. Este enfoque tiene como propósito intuir posibles relaciones entre las variables y buscar otras formas de rellenar los valores faltantes.


```python
# Calcular la matriz de correlación
corr_matrix = df.corr()

# Crear un mapa de calor para visualizar la matriz de correlación
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Matriz de Correlación del Dataset')
plt.show()
```

Se ha identificado una significativa correlación entre las variables "total_vaccinations" y "people_vaccinated". Con el propósito de abordar los datos faltantes en "total_vaccinations" y en última instancia, buscar alternativas para solucionar los faltantes en "people_vaccinated", se ha empleado una estrategia específica. Esta estrategia implica rellenar los valores faltantes en "total_vaccinations" utilizando el valor previo de "total_vaccinations" al cual se le suma el número diario de vacunaciones. Se espera que esta metodología resulte efectiva en la eliminación de los valores faltantes en "people_vaccinated".

```python
# Ordenar el DataFrame por país y fecha
df.sort_values(by=['country', 'date'], inplace=True)

# Crear una columna para almacenar los valores corregidos de total_vaccinations
df['corrected_total_vaccinations'] = df['total_vaccinations']

# Iterar a través de las filas con NaN en "total_vaccinations" y realizar la corrección
for index, row in df.iterrows():
    if pd.isnull(row['total_vaccinations']):  # Verificar si el valor es NaN
        previous_total_vaccinations = df[(df['country'] == row['country']) & (df['date'] < row['date'])].tail(1)['corrected_total_vaccinations']
        if not previous_total_vaccinations.empty:  # Verificar que se encontró un valor anterior del mismo país
            df.at[index, 'corrected_total_vaccinations'] = previous_total_vaccinations.iloc[0] + row['daily_vaccinations']  # Actualizar el valor corregido
        else:
            # No se encontraron valores anteriores del mismo país, se mantiene el NaN
            pass

# Reemplazar la columna original "total_vaccinations" con los valores corregidos
df['total_vaccinations'] = df['corrected_total_vaccinations']
df.drop(columns=['corrected_total_vaccinations'], inplace=True)  # Eliminar la columna temporal
# Utiliza el método isna() para identificar los valores faltantes y luego usa el método sum() para contarlos
missing_values = df.isna().sum()

# Muestra los valores faltantes en cada columna
print(missing_values)
```

Basándonos en esta fórmula corroborada, procederemos a aplicarla con el fin de abordar el resto de valores faltantes.

```python
# Reemplazar los faltantes en "people_vaccinated" con la diferencia entre "total_vaccinations" y "people_fully_vaccinated"
df['people_vaccinated'] = df.apply(lambda row: row['total_vaccinations'] - row['people_fully_vaccinated'] if pd.isnull(row['people_vaccinated']) else row['people_vaccinated'], axis=1)

# Imprimir los valores que se han reemplazado
replaced_values = df[pd.isnull(df['people_vaccinated'])][['total_vaccinations', 'people_fully_vaccinated']]
print(replaced_values)
```

De esta forma bajo la suposición de que la segunda dosis se recibe a los 15 días de haber recibido la primera dósis se pondrán a 0 aquellos valores que sean NA en los primeros 15 días de la vacunación

```python
# Establecer el valor de "people_fully_vaccinated" como 0 para los primeros 15 días de vacunación si el valor es NaN
for country in df['country'].unique():
    mask = (df['country'] == country)
    first_15_days_nan = df[mask]['people_fully_vaccinated'].head(15).isnull()
    df.loc[mask & first_15_days_nan, 'people_fully_vaccinated'] = 0
```

Y por esa variación en la gente completamente vacunada se variarán los valores pertinentes en la gente vacunada con la fórmula vista gráficamente en el anterior apartado

Ahora despejando buscamos encontrar a aquellos valores faltantes que se encuentren en la gente que ha sido vacunada completamente. Esto lo realizamos con el fin de completar la base de datos haciendo que sea más fiable y fácil de trabajar.

```python
# Reemplazar faltantes en "people_fully_vaccinated" con la diferencia entre "total_vaccinations" y "people_vaccinated"
df['people_fully_vaccinated'].fillna(df['total_vaccinations'] - df['people_vaccinated'], inplace=True)
# Utiliza el método isna() para identificar los valores faltantes y luego usa el método sum() para contarlos
missing_values = df.isna().sum()

# Muestra los valores faltantes en cada columna
print(missing_values)
```

Al concluir este ejercicio, se destaca la influencia de la variación de una columna en la posibilidad de que se modifique nuevamente la columna que previamente ajustamos. Este fenómeno indica la interdependencia de los valores y sugiere la necesidad de una solución iterativa para abordar de manera integral los valores faltantes en los datos.

La justificación para este enfoque radica en la interrelación dinámica entre las distintas columnas de datos, lo que sugiere que la actualización de un valor puede afectar directa o indirectamente a otros valores dentro del conjunto de datos. Por lo tanto, al iterar a través de un bucle for, se espera lograr la eliminación progresiva y sistemática de los valores faltantes en todas las columnas pertinentes.

Esta estrategia iterativa permitirá un enfoque completo para abordar los valores faltantes, asegurando que los datos estén completamente actualizados y coherentes en todas las columnas relevantes. Además, al adoptar un enfoque que considera la relación entre las columnas, se promueve la integridad y precisión en el conjunto de datos generado.

```python
# Convertir la columna 'date' a tipo datetime
df['date'] = pd.to_datetime(df['date'])

# Crear una nueva columna para almacenar los valores corregidos
df['corrected_people_vaccinated'] = df['people_vaccinated']
df['corrected_total_vaccinations'] = df['total_vaccinations']
df['corrected_fully_vaccinated'] = df['people_fully_vaccinated']

# Iterar a través de las filas y realizar las correcciones
for index, row in df.iterrows():
    if pd.isnull(row['people_vaccinated']):
        previous_dates = df[(df['country'] == row['country']) & (df['date'] < row['date'])]['date']
        if not previous_dates.empty:
            previous_vaccinated = df[(df['country'] == row['country']) & (df['date'].isin(previous_dates))]['people_vaccinated']
            if not previous_vaccinated.empty:
                mean_previous_vaccinated = np.round(previous_vaccinated.mean())
                df.at[index, 'people_vaccinated'] = mean_previous_vaccinated

    if pd.isnull(row['total_vaccinations']):
        previous_total_vaccinations = df[(df['country'] == row['country']) & (df['date'] < row['date'])].tail(1)['corrected_total_vaccinations']
        if not previous_total_vaccinations.empty:
            df.at[index, 'corrected_total_vaccinations'] = previous_total_vaccinations.iloc[0] + row['daily_vaccinations']

    if pd.isnull(row['people_fully_vaccinated']):
        country_data = df[df['country'] == row['country']]
        prev_vaccinations = country_data[country_data['date'] == row['date'] - pd.Timedelta(days=15)]['people_vaccinated'].values
        prev_fully_vaccinated = country_data[country_data['date'] == row['date'] - pd.Timedelta(days=1)]['corrected_fully_vaccinated'].values
        if prev_vaccinations.size > 0 and prev_fully_vaccinated.size > 0:
            increment = prev_vaccinations[0] - prev_fully_vaccinated[0]
            df.at[index, 'corrected_fully_vaccinated'] = prev_fully_vaccinated[0] + increment

# Reemplazar los valores corregidos en el DataFrame
df['total_vaccinations'] = df['corrected_total_vaccinations']
df['people_vaccinated'] = df.apply(lambda row: row['total_vaccinations'] - row['people_fully_vaccinated'] if pd.isnull(row['people_vaccinated']) else row['people_vaccinated'], axis=1)
df['people_fully_vaccinated'].fillna(df['total_vaccinations'] - df['people_vaccinated'], inplace=True)

# Establecer el valor de "people_fully_vaccinated" como 0 para los primeros 15 días de vacunación si el valor es NaN
for country in df['country'].unique():
    mask = (df['country'] == country)
    first_15_days_nan = df[mask]['people_fully_vaccinated'].head(15).isnull()
    df.loc[mask & first_15_days_nan, 'people_fully_vaccinated'] = 0

# Ordenar el DataFrame por país y fecha
df.sort_values(by=['country', 'date'], inplace=True)

# Eliminar las columnas temporales
df.drop(columns=['corrected_total_vaccinations', 'corrected_people_vaccinated', 'corrected_fully_vaccinated'], inplace=True)
```

Finalmente, se observa como han desaparecido ya todos los valores faltantes en los campos deseados  

```python
# Utiliza el método isna() para identificar los valores faltantes y luego usa el método sum() para contarlos
missing_values = df.isna().sum()

# Muestra los valores faltantes en cada columna
print(missing_values)
```

```python
# Graficar people_vaccinated_per_hundred
for country in df["country"].unique():
    plt.plot(df[df["country"] == country]["date"], df[df["country"] == country]["people_vaccinated_per_hundred"], label=country)
plt.xlabel('Fecha')
plt.ylabel('People Vaccinated per Hundred')
plt.title('People Vaccinated per Hundred over Time')
plt.legend()
plt.show()
# Graficar total_vaccinations_per_hundred
for country in df["country"].unique():
    plt.plot(df[df["country"] == country]["date"], df[df["country"] == country]["total_vaccinations_per_hundred"], label=country)
plt.xlabel('Fecha')
plt.ylabel('Total Vaccinations per Hundred')
plt.title('Total Vaccinations per Hundred over Time')
plt.legend()
plt.show()

# Graficar people_fully_vaccinated_per_hundred
for country in df["country"].unique():
    plt.plot(df[df["country"] == country]["date"], df[df["country"] == country]["people_fully_vaccinated_per_hundred"], label=country)
plt.xlabel('Fecha')
plt.ylabel('People Fully Vaccinated per Hundred')
plt.title('People Fully Vaccinated per Hundred over Time')
plt.legend()
plt.show()

# Graficar daily_vaccinations_per_million
for country in df["country"].unique():
    plt.plot(df[df["country"] == country]["date"], df[df["country"] == country]["daily_vaccinations_per_million"], label=country)
plt.xlabel('Fecha')
plt.ylabel('Daily Vaccinations per Million')
plt.title('Daily Vaccinations per Million over Time')
plt.legend()
plt.show()


```

```python
poblacion = {
    'United Kingdom': 68000000,
    'Israel': 8655000,
    'Russia': 145912025,
    'China': 1444216107,
    'United States': 331002651
}

# Aplicar transformación per hundred en people_vaccinated_per_hundred para valores faltantes
for country in df["country"].unique():
    if pd.isnull(df.loc[df['country'] == country, 'people_vaccinated_per_hundred']).any():
        df.loc[df['country'] == country, 'people_vaccinated_per_hundred'] = df.loc[df['country'] == country, 'people_vaccinated'] / poblacion[country] * 100

# Aplicar transformación per hundred en total_vaccinations_per_hundred para valores faltantes
for country in df["country"].unique():
    if pd.isnull(df.loc[df['country'] == country, 'total_vaccinations_per_hundred']).any():
        df.loc[df['country'] == country, 'total_vaccinations_per_hundred'] = df.loc[df['country'] == country, 'total_vaccinations'] / poblacion[country] * 100

# Aplicar transformación per hundred en people_fully_vaccinated_per_hundred para valores faltantes
for country in df["country"].unique():
    if pd.isnull(df.loc[df['country'] == country, 'people_fully_vaccinated_per_hundred']).any():
        df.loc[df['country'] == country, 'people_fully_vaccinated_per_hundred'] = df.loc[df['country'] == country, 'people_fully_vaccinated'] / poblacion[country] * 100

# Aplicar transformación per million en daily_vaccinations_per_million para valores faltantes
for country in df["country"].unique():
    if pd.isnull(df.loc[df['country'] == country, 'daily_vaccinations_per_million']).any():
        df.loc[df['country'] == country, 'daily_vaccinations_per_million'] = df.loc[df['country'] == country, 'daily_vaccinations'] / poblacion[country] * 1000000

```

Para evaluar correctamente un diagrama de violín se debe tener en cuenta:


*   Forma de violín: La forma de violín en el gráfico representa la densidad de probabilidad de los datos. Un violín más ancho indica una mayor densidad de datos en ese rango.

*   Caja y bigotes: La "caja" en el centro del violín representa el rango intercuartíl (IQR) de los datos, similar a un diagrama de caja y bigotes. La línea en el centro de la caja indica la mediana de los datos.

*   Extremos del violín: Los extremos de los "violines" muestran la distribución de los datos más allá de los valores mínimos y máximos. Pueden ayudarte a observar la concentración y dispersión de los datos.

*   Comparación: Los diagramas de violín también son útiles para comparar las distribuciones de varios grupos de datos en un solo gráfico.

La matriz de covarianza es una herramienta fundamental en el análisis estadístico que permite comprender la relación y la dispersión conjunta de múltiples variables en un conjunto de datos. Ayuda a describir cómo cada par de variables se mueve en relación con el otro, proporcionando información sobre la dirección y la fuerza de la relación lineal entre ellas. En particular, la matriz de covarianza es fundamental en el análisis de componentes principales (PCA) para determinar la estructura de correlación y la cantidad de información contenida en las variables originales.

Al calcularla matriz de covarianza de un conjunto de datos, se obtiene un cuadro de doble entrada que muestra la varianza de cada variable en la diagonal principal, y las covarianzas entre pares de variables en las entradas fuera de la diagonal. Esta información es crucial para comprender la magnitud y la dirección de la relación entre variables, lo cual es esencial para el posterior cálculo de los autovalores y autovectores que caracterizan los componentes principales de los datos.

```python
# Calcular la matriz de covarianza
cov_matrix = np.cov(data_standardized, rowvar=False)
```

La función linalg.eig en NumPy se utiliza para calcular los autovalores y autovectores de una matriz cuadrada, lo que es fundamental en el análisis de componentes principales (PCA) y en diversos procedimientos de álisis y transformación de datos.

Al aplicar linalg.eig a la matriz de covarianza previamente calculada, se obtienen los autovalores y autovectores asociados. Los autovalores representan la cantidad de varianza explicada por cada componente principal, mientras que los autovectores son los vectores propios que describen la dirección de máxima varianza de los datos en un nuevo sistema de coordenadas.

Estos autovalores y autovectores proporcionan información crítica sobre la estructura de correlación de los datos y la cantidad de información contenida en cada componente principal. Al calcular el porcentaje de varianza explicada y el porcentaje acumulado por cada componente, se puede determinar cuánta información está contenida en cada componente y cuánto se acumula al considerar múltiples componentes.


```python
# Calcular autovalores y autovectores
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
# Ordenar los eigenvalues y eigenvectors
sorted_indices = np.argsort(eigenvalues)[::-1]  # Obtener los índices que ordenarían los eigenvalues de forma descendente
eigenvalues_sorted = eigenvalues[sorted_indices]  # Ordenar los eigenvalues
eigenvectors_sorted = eigenvectors[:, sorted_indices]  # Ordenar los eigenvectors según los eigenvalues

# Mostrar los eigenvalues y eigenvectors ordenados
print("Eigenvalues ordenados:")
print(eigenvalues_sorted)
print("\nEigenvectors ordenados:")
print(eigenvectors_sorted)
```

A continuación, se procede a calcular el porcentaje de varianza explicada por cada componente. Este cálculo permitirá ordenar los componentes de manera descendente en función de su varianza, lo que a su vez facilitará el análisis sobre cuántos componentes deben ser considerados para representar adecuadamente la estructura de los datos. Además, esta evaluación proporcionará información crítica sobre cuánta información se perdería si se seleccionara un número insuficiente de componentes, lo que es esencial para comprender el impacto de la reducción de dimensionalidad en la retención de información importante en el conjunto de datos.

```python
# Calcular el porcentaje de varianza explicada por cada componente
explained_variance_ratio = eigenvalues / sum(eigenvalues)
explained_variance_ratio = sorted(explained_variance_ratio, reverse=True)
explained_variance_ratio_cumulative = np.cumsum(explained_variance_ratio)
explained_variance_ratio_cumulative
```

Basado en los resultados obtenidos al calcular el porcentaje de varianza explicada por cada componente, y considerando la varianza explicada acumulada, se ha tomado la decisión de realizar el análisis con 3 componentes principales.

Al ordenar los porcentajes de varianza explicada de mayor a menor, se observa que los tres primeros componentes capturan un porcentaje significativo de la varianza, acumulando más del 80% de la varianza explicada.

Al seleccionar 3 componentes principales, se retiene una cantidad substancial de información, lo cual resulta en la representación efectiva de la estructura subyacente de los datos originales. Esta decisión se basa en el equilibrio entre la retención de información importante y la simplificación del modelo, lo que facilitará la comprensión de los datos y la extracción de patrones relevantes para el análisis subsiguiente.

Por esto:

```python
print('Si se cojen 2 componentes únicamente se pierde un', (1-explained_variance_ratio_cumulative[1]), 'de la información')
```

El Análisis de Componentes Principales (PCA) es una técnica estadística utilizada para comprimir la información contenida en un conjunto de variables correlacionadas en un conjunto más pequeño de variables no correlacionadas llamadas componentes principales. La finalidad principal de PCA es reducir la dimensionalidad de los datos al mismo tiempo que se mantiene la mayor cantidad de información posible, lo que facilita la visualización y análisis de conjuntos de datos complejos en un espacio de menor dimensión.

Los pesos en un Análisis de Componentes Principales (PCA) representan la magnitud de la contribución de cada variable original a los componentes principales derivados. Estos pesos nos brindan información sobre cómo cada variable original impacta en la formación de los componentes principales. Al examinar los pesos, podemos comprender qué variables tienen una mayor influencia en cada componente principal, lo que a su vez nos proporciona insights sobre la estructura subyacente de los datos y las relaciones entre las variables originales.

Al aplicar PCA y examinar los pesos asociados a cada variable en los componentespales, podemos entender cómo las variables originales contribuyen a la estructura subyacente de los datos. Esto nos ayuda a interpretar y comprender la naturaleza de cada componente principal, lo que a su vez proporciona información sobre qué aspectos o patrones de los datos son capturados por cada componente. De esta manera, los pesos en un PCA son fundamentales para entender y explicar qué representa cada columna en el conjunto de datos reducido resultante del análisis.

```python
pca = PCA(n_components=3)
principal_components = pca.fit_transform(data_standardized)

# Obtener los pesos del PCA
weights = pca.components_
# Crear un DataFrame para la matriz de pesos con los nombres de las variables originales como índices
# Convertir la matriz a un DataFrame
weights_df = pd.DataFrame(weights, columns=nueva_data.columns, index=['Component 1', 'Component 2', 'Component 3'])  # Reemplaza 'Component 1', 'Component 2', 'Component 3' con los nombres de tus componentes

# Mostrar los pesos del PCA
print(weights_df)


# Crear un diagrama de dispersión en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(principal_components[:, 0], principal_components[:, 1], principal_components[:, 2])

# Etiquetar las ciudades en el gráfico
for i, city in enumerate(data.index):
    ax.text(principal_components[i, 0], principal_components[i, 1], principal_components[i, 2], city)

# Configurar etiquetas de los ejes
ax.set_xlabel('Componente Principal 1')
ax.set_ylabel('Componente Principal 2')
ax.set_zlabel('Componente Principal 3')

plt.show()

```
