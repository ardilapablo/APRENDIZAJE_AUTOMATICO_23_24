El conjunto de datos está diseñado para el reconocimiento de objetos y la clasificación de imágenes en categorías muy detalladas. Fue recopilado por Alex Krizhevsky, Vinod Nair y Geoffrey Hinton, y publicado por la Universidad de Toronto.

El propósito de este conjunto de datos es desafiar los algoritmos de aprendizaje automático a reconocer e interpretar una amplia variedad de objetos. Es un recurso importante para entrenar modelos de aprendizaje automático en tareas de clasificación de imágenes.

Al cargar el conjunto de datos CIFAR-100 en Python con Keras, obtendrás acceso a las 60,000 imágenes para el entrenamiento y 10,000 imágenes para evaluar la precisión del modelo. Dado que CIFAR-100 tiene 100 clases diferentes, la función to_categorical convierte cada etiqueta de clase en un vector binario de longitud 100. Por ejemplo, si originalmente teníamos una etiqueta como 3, al aplicar to_categorical esta se convertiría en un vector binario con un 1 en la cuarta posición y 0 en todas las demás posiciones:

[0, 0, 0, 1, 0, ..., 0]

Cada clase se representa como un vector binario para que el modelo de aprendizaje automático pueda interpretar las etiquetas de forma más efectiva durante el entrenamiento y la evaluación. Este paso es común al trabajar con conjuntos de datos que contienen múltiples clases y es fundamental para la efectividad de nuestro modelo de clasificación.

```python
#ytr = ytr.reshape(-1,)
#yt = yt.reshape(-1,)
ytr = to_categorical(ytr, 100)
yt = to_categorical(yt, 100)
```

## Preparación de los datos y Creación de la red

 La normalización de los datos es un paso importante al trabajar con nuestro modelo de aprendizaje automático, incluso en redes neuronales convolucionales (CNN) como las que se utilizan en el procesamiento de imágenes. La normalización de los datos significa ajustar los valores de entrada para que tengan una escala común, lo que puede ayudar a mejorar la convergencia del modelo y su rendimiento.

En el contexto de procesamiento de imágenes, los píxeles de las imágenes suelen tener valores que van desde 0 hasta 255, ya que representan la intensidad de los canales de color (rojo, verde, azul). La normalización típica para las imágenes consiste en escalar estos valores para que estén en el rango de 0 a 1. Esto se logra dividiendo cada valor de píxel por 255.

```python
xtr = xtr / 255
xt = xt / 255
```

La división de los datos en un conjunto de entrenamiento y un conjunto de validación es esencial para evaluar el rendimiento y evitar el sobreajuste. Durante el entrenamiento, el conjunto de datos de entrenamiento se utiliza para ajustar los pesos, mientras que el conjunto de datos de validación se usa para evaluar el rendimiento en datos no vistos y tomar decisiones sobre la arquitectura y los hiperparámetros.

En cada iteración del entrenamiento, después de ajustar los datos de entrenamiento, se evalúa el rendimiento utilizando los datos de validación. Esto proporciona una estimación del rendimiento en datos no vistos y facilita la detección temprana de problemas como el sobreajuste.

Es fundamental supervisar cómo se desempeña el modelo en el conjunto de validación durante el entrenamiento, ya que esto permite tomar decisiones importantes sobre los ajustes en la arquitectura del modelo, la tasa de aprendizaje y otros hiperparámetros. Ajustar estos hiperparámetros en base al rendimiento en el conjunto de validación tiene como objetivo mejorar la capacidad predictiva del modelo y prevenir que se sobreajuste a los datos de entrenamiento.

```python
x_val = xtr[:10000]
pxtr = xtr[10000:]
y_val = ytr[:10000]
pytr = ytr[10000:]

```

Para modelar el conjunto de datos CIFAR-100 con una red neuronal, se ha diseñado una arquitectura que busca extraer características relevantes de las imágenes y aprender a clasificarlas en las 100 categorías distintas de forma precisa. La justificación para cada parte de la arquitectura es la siguiente:

Capas convolucionales: Las capas convolucionales son fundamentales para capturar patrones locales en las imágenes, y su utilización en esta red busca identificar características útiles para la clasificación. El uso de 128 filtros de 3x3 ayuda a la red a aprender una variedad de características a diferentes escalas en las imágenes.

Función de Activación 'elu': La función de activación 'elu' se utiliza para introducir no linealidades en la red, permitiendo la modelación de comportamientos más complejos en los datos de entrada. Esta no linealidad es crucial para que la red pueda aprender y representar de manera adecuada la información.

Capas de MaxPooling: La inserción de capas de MaxPooling con un tamaño de ventana de 2x2 tiene como objetivo reducir la dimensión de la salida de las capas convolucionales, lo que permite reducir la complejidad de la red y el cálculo computacional necesario, al mismo tiempo que preserva las características más importantes.

Capas de Dropout: El Dropout es una técnica de regularización que se utiliza para prevenir el sobreajuste de la red. Al eliminar aleatoriamente conexiones entre las neuronas durante el entrenamiento, se evita que la red dependa en exceso de ciertas características o patrones en los datos. Esto ayuda a que la red sea más robusta y generalice mejor a datos no vistos.

Capa Flatten: Después de las capas convolucionales y de reducción de dimensionalidad, se utiliza una capa Flatten para transformar la salida en un vector unidimensional. Esto prepara los datos para ser introducidos en la capa densamente conectada.

Capas Densamente Conectadas: Las capas densamente conectadas ayudan a combinar las características extraídas por las capas convolucionales y aprender representaciones más abstractas de las imágenes. En este caso, la presencia de 1024 neuronas y la posterior aplicación de Dropout ayudan a capturar y complementar información relevante para la clasificación.

```python
#añadimos capas secuenciales (una detrás de la otra)
model = Sequential()
#añadimos una capa convolucional con 128 filtros de 3x3. Hemos comprobado que
#menos filtros bajan la precisión considerablmente
model.add(Conv2D(128, (3, 3), padding='same', input_shape=xtr.shape[1:]))
model.add(Activation('elu'))
model.add(Conv2D(128, (3, 3)))#keras deducirá el input_shape a partir de ahora
model.add(Activation('elu'))
#añadimos una capa de MaxPooling para ir reduciendo la dimensión de la salida.
#una capa más grande provocaría demasiada pérdida de información
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(256, (3, 3), padding='same'))
model.add(Activation('elu'))

model.add(Conv2D(256, (3, 3)))
model.add(Activation('elu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
#añadimos también una capa de dropout para evitar el sobreentrenamiento.
#Las capas dropout convierten a 0 datos aleatorios.
model.add(Dropout(0.25))

model.add(Conv2D(512, (3, 3)))
model.add(Activation('elu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
#añadimos también una capa flatten que transformará la salida de las capas anteriores
#en un vector de una sola dimensión
model.add(Flatten())
model.add(Dense(1024))
model.add(Activation('elu'))
model.add(Dropout(0.5))
#acabamos con una capa dense con 100 neuronas (número de categorías) y función
#de activación softmax
model.add(Dense(100))
model.add(Activation('softmax'))
```

```python
#Vamos a hacer data augmentation:
datagen = ImageDataGenerator(
    featurewise_center=False,#no se tiene como referencia la media de todo el
    #conjunto de datos
    samplewise_center=False,
    featurewise_std_normalization=False,#normalizar implicaría dividir cada
    #dato por la desviación estándar, o sea que nos lo ahorramos
    samplewise_std_normalization=False,
    zca_whitening=False,
    rotation_range=0,
    width_shift_range=0.1,#la imagen se desplaza hasta un 10% en el eje horizintal
    height_shift_range=0.1,#igual pero en el eje vertical
    horizontal_flip=True,#se dan la vuelta a algunas imágenes aleatoriamente horizontalmente
    vertical_flip=False
)
```

```python
#vamos a usar el optimizador Adam, que ajusta individualmente la tasa de aprendizaje
#para cada parámetro de la red
optimizer = Adam(learning_rate=0.0001, weight_decay=1e-6)
#el weight_decay ayudará a prevenir el sobreentrenamiento
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
#accuracy medirá el porcentaje de predicciones correctas
```

```python
#comenzamos el proceso de entrenamiento con los datos de entrenamiento sin los
#validación
#se utilza para la normalizar y aumentar los datos durante el entrenamiento de nuestra red
#mejorará la aumentación de datos en tiempo real
datagen.fit(pxtr)
```

```python
#EarlyStopping nos permitirá detener el entrenamiento antes en caso de que no
#haya mejoras significativas en el conjunto de validación
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=8, restore_best_weights = True)

#Entrenaremos el modelo usando el generador de datos
history = model.fit(datagen.flow(pxtr, pytr, batch_size=50),
                   #el número de pasos por época será el número de muestrás
                   #dividido por el batch_size
                   steps_per_epoch=pytr.shape[0] // 50,
                   epochs=80,
                   validation_data=(x_val, y_val),
                   #shuffle mezcla los datos en cada época, antes de dividirlos en
                   #lotes
                   shuffle = True,
                   #verbose = 1 provoca que se muestre información más detallada
                   #durante el entrenamiento
                   verbose=1,
                   callbacks=[early_stop])
```

```python
test_loss, test_acc = model.evaluate(xt, yt)
print("test accuracy: ", test_acc)
```

```python
history_dict = history.history
history_dict.keys()
print(history_dict['loss'])
print(history_dict['accuracy'])
print(history_dict['val_loss'])
print(history_dict['val_accuracy'])
```
